```python
import pandas as pd
import PIL 
import os 
import pathlib
from animal_classification.data import AnimalDataSet

data_dir = "../data/raw/raw-img" 
files = list(pathlib.Path(data_dir).glob("*/*.jpeg"))
```

```python
import matplotlib.pyplot as plt 


fig,ax = plt.subplots(4,4, figsize=(10,10)) 

N = 16

for idx in range(N): 
    row, col = divmod(idx, 4)
    img = PIL.Image.open(files[idx])
    ax[row, col].imshow(img)

plt.show()
```
